```python
import torchaudio
import glob

sample_rate_target = 16000

audio_items = []   # list of dicts: {name, label, wav}

def load_folder(folder, label):
    files = sorted(glob.glob(folder + "/*.mp3"))
    print(f"Found {len(files)} files in {label}.")
    for p in files:
        wav, sr = torchaudio.load(p)
        if wav.shape[0] > 1:
            wav = wav.mean(dim=0, keepdim=True)
        wav = wav.squeeze(0)
        if sr != sample_rate_target:
            wav = torchaudio.functional.resample(wav, sr, sample_rate_target)
        audio_items.append({
            "name": os.path.basename(p),
            "label": label,
            "wav": wav,
        })

load_folder(mainstream_dst, "mainstream")
load_folder(non_mainstream_dst, "non_mainstream")

print("Total audio loaded:", len(audio_items))
```

```python
import torch
import pandas as pd

zipa_model.eval()

def extract_ipa(output):
    """
    Flatten nested list/tuple structure until we get a single string.
    Works for things like ['ipa'], [['ipa']], etc.
    """
    x = output
    while isinstance(x, (list, tuple)):
        if not x:          # empty list/tuple
            return ""
        x = x[0]
    return str(x)

rows = []
with torch.no_grad():
    for item in audio_items:
        raw = zipa_model.inference([item["wav"]])
        ipa = extract_ipa(raw)   # <- guaranteed to be a string

        rows.append({
            "audio_file": item["name"],   # e.g. "common_voice_it_25595019.mp3"
            "ipa_zipa": ipa               # plain string, no brackets
        })

df_zipa = pd.DataFrame(rows)
df_zipa.head()
```

```python
def segment_ipa_phones(s):
    """
    Simple IPA phone segmentation.
    Treats tʃ, dʒ, ts, dz as single units.
    Returns a list of 'phones' as strings.
    """
    # Just in case something weird comes in
    if isinstance(s, list):
        s = s[0] if s else ""
    if not isinstance(s, str):
        s = str(s)

    # use placeholders for multi-character phones
    s = (s.replace("tʃ", "ƭ")
           .replace("dʒ", "ƈ")
           .replace("ts", "Ʈ")
           .replace("dz", "Ɗ"))

    phones = []
    for ch in s:
        if ch == "ƭ":
            phones.append("tʃ")
        elif ch == "ƈ":
            phones.append("dʒ")
        elif ch == "Ʈ":
            phones.append("ts")
        elif ch == "Ɗ":
            phones.append("dz")
        else:
            phones.append(ch)
    return phones

df_zipa["ipa_zipa_phones"] = df_zipa["ipa_zipa"].apply(
    lambda s: " ".join(segment_ipa_phones(s))
)

df_zipa
```
